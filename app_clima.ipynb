```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt

def extrae(nombre_archivo, ten_header=True):
    """
    Lee un archivo CSV y devuelve un DataFrame de pandas.

    Parmetros:
    - nombre_archivo: str. Ruta del archivo CSV.
    - ten_header: bool. True si el archivo tiene encabezado, False si no.

    Retorna:
    - DataFrame con los datos.
    """
    if ten_header:
        df = pd.read_csv(nombre_archivo, header=0)
    else:
        df = pd.read_csv(nombre_archivo, header=None)
    return df

# === Configuracin inicial ===
nombre_archivo = "/content/temperatura_ejemplo_24h.csv"     # Cambia esto por la ruta a tu archivo
ten_header = True                # True si tu CSV tiene encabezado
formato_fecha = "%Y/%m/%d %H:%M:%S"  # Ajusta si tu formato de fecha es diferente

# === Lectura del archivo CSV ===
datos = extrae(nombre_archivo, ten_header)

# === Procesamiento de fechas ===
x = datos["hora"]                # Columna de hora
fechas = []

for time in x:
    temp = datetime.datetime.strptime(time, formato_fecha)
    fechas.append(temp)

# === Extraccin de los valores de temperatura ===
temperatura = datos["T"]         # Columna de temperatura en C

# === Grfica ===
plt.figure(figsize=(10, 5))
plt.plot(fechas, temperatura, color='tomato', marker='o', linestyle='-')
plt.xlabel("Fecha y hora")
plt.ylabel("Temperatura (C)")
plt.title("Temperatura vs Tiempo")
plt.grid(True)
plt.tight_layout()
plt.xticks(rotation=45)
plt.show()
```
